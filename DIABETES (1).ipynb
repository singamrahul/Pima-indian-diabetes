```python
data.columns = ['pregnancies' , 'glucose' , 'blood pressure' , 'skin thickness' , 'insulin' , 'bmi' , 'diabetes pedigree function' , 'age' , 'outcome']
```

```python
data  # assigned new column names
```

```python
# to check are there any null values present in the dataset
data.isnull().sum()
```

```python
'''
Technically there are no null values present in the dataset but practically there are some columns where min value
starts with 0 which was not possible in real life scenario so we need to assume the as null values and handle it
'''
```

```python
# actually i thought to delete the rows in these condition but rows are limited to instead of removing the values i have
# imputed the values with median bcoz median is not sensitive to outliers
```

```python
features = data.columns
cols = (data[features] == 0).sum()
cols
```

```python
data[['glucose','blood pressure','skin thickness','insulin','bmi']] = data[['glucose','blood pressure','skin thickness','insulin','bmi']].replace(0,np.NaN)
```

```python
data.isnull().sum()
```

```python
data['glucose'].fillna(data['glucose'].median(), inplace =True)
data['blood pressure'].fillna(data['blood pressure'].median(), inplace =True)
data['bmi'].fillna(data['bmi'].median(), inplace =True)
data['skin thickness'].fillna(data['skin thickness'].median(),inplace = True)
data['insulin'].fillna(data['insulin'].median(),inplace = True)
```
